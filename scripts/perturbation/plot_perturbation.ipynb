```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import matplotlib.ticker as ticker
from scipy.stats import kstest
from matplotlib.ticker import MaxNLocator


for fname in ['MNIST-5mlp-perturb-minus', 'MNIST-5mlp-perturb']:
    for suff in [ ['mlp','MNIST-5mlp']]:
        layer_list = ['all']
        labelsize=20
        for opti in ['PureSGD']:
            lr_list = [0.01]
            pertb_epochs = 5000
            colors = plt.cm.viridis(np.linspace(0,1,4))
            plt.figure(figsize=(6,5))

            for i, lr in enumerate(lr_list):
                for st_ep in [50, 100]:
                    
                    if st_ep == 50:
                        mark = 'o'
                        color = colors[0]
                    elif st_ep == 100:
                        mark = 's'
                        color = colors[1]

                    for (vi, value) in enumerate([0.01]):
                           
                        all_loss = np.load(f'{fname}_ep{st_ep}_perturb{value}_lr{lr}_opti{opti}_lossvariance.npy')
                        if fname == 'MNIST-5mlp-perturb-minus':
                            nums, bins = np.histogram(all_loss, bins=40)
                        else:
                            nums, bins = np.histogram(all_loss, bins=40)
                        nums = nums/np.sum(nums)
                        xbins = bins[:-1] + np.diff(bins)/2
                        lal = 'All Layers'

                        plt.semilogy(xbins, nums, mark, markersize=12, alpha=0.9, color=color, label=f'perturbed at {st_ep} epoch')

                        plt.xlabel(r'$|\Delta L|$',fontsize=labelsize)
                        plt.ylabel('Probability', fontsize=labelsize)
                

                plt.gca().xaxis.set_major_formatter(ticker.ScalarFormatter(useMathText=True))
                plt.gca().xaxis.get_major_formatter().set_powerlimits((-2, 2))  
                ax = plt.gca()
                ax.xaxis.offsetText.set_fontsize(labelsize-2) 
                ax.yaxis.offsetText.set_fontsize(labelsize-2)  
                ax.xaxis.set_major_locator(MaxNLocator(nbins=5))

                plt.xticks(fontsize=labelsize)
                plt.yticks(fontsize=labelsize)
                if fname == 'MNIST-5mlp-perturb-minus':
                    plt.title('Under negative perturbation\n($\\delta_p = -0.01$)', fontsize=labelsize)
                else:
                    plt.title('Under positive perturbation\n($\delta_p=0.01$)', fontsize=labelsize)
                plt.legend(fontsize=labelsize-5, frameon=False, loc='lower left')
                plt.show()

                
```
